```python
from numpy import True_


nums = []
file = open("out.txt")
for line in file.readlines():
    try:
        nums.append(float(line))
    except:
        continue


nums.sort(reverse=True)
len(nums)
```

```python
yes = [nums[0]/nums[i] for i in range(len(nums))]
xes = [i for i in range(len(nums))]
```

```python
import matplotlib.pyplot as plt

plt.plot(xes, yes)
plt.scatter(xes, yes, color='orange', s=20, marker='o')
plt.title("ускорение")
plt.ylabel("t[0]/t[x]")
plt.xlabel("cores")
plt.grid()
plt.show()
```
