```python
#transfer node
from numpy import float64, power, array

x1 = []
y1 = []
with open("one_node.txt") as src:
    for i in src.readlines():
        i = i.split(",")
        x1.append(float64(i[1]))
        y1.append(float64(i[0]))
x1, y1 = array(x1), array(y1)
```

```python
#two nodes
x2 = []
y2 = []
with open("different_nodes.txt") as src:
    for i in src.readlines():
        i = i.split(",")
        x2.append(float64(i[1]))
        y2.append(float64(i[0]))
x2, y2 = array(x2), array(y2)
```

```python
import matplotlib.pyplot as plt

transfer = plt.subplot()
transfer.plot(x1[22:], y1[22:], x2[22:], y2[22:])
transfer.grid()
transfer.set_title("transfer message")


```

```python
alpha = sum([y2[i] for i in range(15)])/15
betta = (y2[30]-y2[29]) / (2**30 - 2**29)
alpha, 1/betta
```

```python

network = plt.subplot()
xes1 = power(2, x1)
yes1 = xes1 / y1
xes2 = power(2, x2)
yes2 = xes2 / y2
network.plot(xes1, yes1, xes2, yes2)
network.grid()
network.set_title("network bandwidth")
```
